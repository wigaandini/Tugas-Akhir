## Quick sanity checks

```python
# Expected: 7 gestures × 3 repetitions × 11 positions × sum_of_sessions trials
# With sessions = 10+10+10+10+10+10 + 3+3+3+3+3+3+3+3 + 1×(36-14) = 60 + 24 + 22 = 106 sessions
# Expected trials = 7 × 3 × 11 × 106 = 24,486 (matches paper)
# Fatigue subjects (h30-h35) also have 11 "positions" in the same structure

print(f"Trials per (subject, session) - should be 231 for standard, same for FA:")
per_session = metadata.groupby(["subject", "session"]).size()
print(per_session.describe())
print()
print("Any sessions with != 231 trials?")
print(per_session[per_session != 231])
```
